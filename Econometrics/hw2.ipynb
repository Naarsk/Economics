```python
import statsmodels.api as sm
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cornwell.csv")
```

```python
df.describe()
```

```python
# filter data to only include year 87
df_87 = df[df['year'] == 87]

# define the variables
X = np.log(df_87[['prbarr', 'prbconv', 'prbpris', 'avgsen']])
y = np.log(df_87['crmrte'])

# add a constant to the independent value
X = sm.add_constant(X)

# fit the model
model = sm.OLS(y, X).fit(cov_type="HC1")

# print the results
print(model.summary())
```

```python
# add the log(crmrate) of year 86 as a regressor
df_86 = df[df['year'] == 86]
X_new = np.column_stack((X, np.log(df_86['crmrte'])))

# add a constant to the independent value
X_new = sm.add_constant(X_new)

# fit the OLS model
model = sm.OLS(y, X_new).fit(cov_type="HC1")

# print the results
print(model.summary())
```
