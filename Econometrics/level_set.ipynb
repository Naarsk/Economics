```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x1, x2, a, b, alpha):
    return (np.minimum(a*x1, b*x2))**alpha

# Create a grid of x1 and x2 values
x1 = np.linspace(0, 10, 100)
x2 = np.linspace(0, 10, 100)
X1, X2 = np.meshgrid(x1, x2)

# Evaluate the function on the grid
Y = f(X1, X2, a=2, b=1, alpha=0.3)

# Plot the contour set
plt.contour(X1, X2, Y, levels=20)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Contour set for y=min(x1, x2)')
plt.show()
```
