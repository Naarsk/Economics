```python
from MacroModels.solow_model import SolowModel, CobbDouglas

cobb_douglas = CobbDouglas(alpha=0.5)

cobb_douglas(x=[5,2])
```

```python
def logistic_map(x):
    return (1+2.02) * x * (1-x*0.002)

def exponential_growth(x):
    return (1+0.02) * x
```

```python
solow_model = SolowModel(saving_rate=0.3, depreciation_rate=0.02, starting_capital=100, production_function=cobb_douglas, starting_pop_level=100,labor_growth_function=logistic_map)

df=solow_model.evaluate()

print(df.to_string());
```

```python
df[['labor']].plot()
df[['income']].plot()
df[['capital']].plot()
df[['wages']].plot()
df[['income_per_cap']].plot()
df[['capital_per_cap']].plot()
df[['rates']].plot()
```
