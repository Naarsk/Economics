```python
from levy_distribution import pdf_from_cf, _psi_tld
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python

# Default parameters
loc = 0
scale = 1
alpha = 0.7
lam = 0.2
beta = -0.2
```

```python


# Create a figure with 5 subplots
fig, axs = plt.subplots(1, 5, figsize=(25, 5), facecolor='white', dpi=200)

# Subplot 1: Varying loc
x_values = np.linspace(-5, 5, 1000)
y_values1 = pdf_from_cf(x_values, loc=0, scale=scale, alpha=alpha, lam=lam, beta=beta, k_max=15.0, ln_2_k_points=12, x_lim=250)
y_values2 = pdf_from_cf(x_values, loc=2, scale=scale, alpha=alpha, lam=lam, beta=beta, k_max=15.0, ln_2_k_points=12, x_lim=250)
axs[0].plot(x_values, y_values1, label='loc=0', color='#C00000', linewidth=1.00)
axs[0].plot(x_values, y_values2, label='loc=2', color='#000000', linewidth=1.00)
axs[0].set_title("Varying loc")
axs[0].grid(color='#E8E8E9', linestyle='-', linewidth=0.8)
axs[0].legend()

# Subplot 2: Varying scale
x_values = np.linspace(-5, 5, 1000)
y_values1 = pdf_from_cf(x_values, loc=loc, scale=1, alpha=alpha, lam=lam, beta=beta, k_max=15.0, ln_2_k_points=12, x_lim=250)
y_values2 = pdf_from_cf(x_values, loc=loc, scale=2, alpha=alpha, lam=lam, beta=beta, k_max=15.0, ln_2_k_points=12, x_lim=250)
axs[1].plot(x_values, y_values1, label='scale=1', color='#C00000', linewidth=1.00)
axs[1].plot(x_values, y_values2, label='scale=2', color='#000000', linewidth=1.00)
axs[1].set_title("Varying scale")
axs[1].grid(color='#E8E8E9', linestyle='-', linewidth=0.8)
axs[1].legend()

# Subplot 3: Varying alpha
x_values = np.linspace(-5, 5, 1000)
y_values1 = pdf_from_cf(x_values, loc=loc, scale=scale, alpha=0.7, lam=lam, beta=beta, k_max=15.0, ln_2_k_points=12, x_lim=250)
y_values2 = pdf_from_cf(x_values, loc=loc, scale=scale, alpha=1.7, lam=lam, beta=beta, k_max=5.0, ln_2_k_points=14, x_lim=250)
y_values3 = norm.pdf(x_values, loc=0, scale=1)
axs[2].plot(x_values, y_values1, label='alpha=0.7', color='#C00000', linewidth=1.00)
axs[2].plot(x_values, y_values2, label='alpha=1.7', color='#000000', linewidth=1.00)
axs[2].plot(x_values, y_values3, label='N(0,1)', color='#008000', linewidth=1.00)
axs[2].set_title("Varying alpha")
axs[2].grid(color='#E8E8E9', linestyle='-', linewidth=0.8)
axs[2].legend()

# Subplot 4: Varying lam
x_values = np.linspace(-5, 5, 1000)
y_values1 = pdf_from_cf(x_values, loc=loc, scale=scale, alpha=alpha, lam=0.1, beta=beta, k_max=15.0, ln_2_k_points=12, x_lim=250)
y_values2 = pdf_from_cf(x_values, loc=loc, scale=scale, alpha=alpha, lam=1, beta=beta, k_max=15.0, ln_2_k_points=12, x_lim=250)
y_values3 = pdf_from_cf(x_values, loc=loc, scale=scale, alpha=alpha, lam=3, beta=beta, k_max=15.0, ln_2_k_points=12, x_lim=250)
axs[3].plot(x_values, y_values1, label='lam=0.1', color='#C00000', linewidth=1.00)
axs[3].plot(x_values, y_values2, label='lam=1', color='#000000', linewidth=1.00)
axs[3].plot(x_values, y_values3, label='lam=3', color='#008000', linewidth=1.00)
axs[3].set_title("Varying lam")
axs[3].grid(color='#E8E8E9', linestyle='-', linewidth=0.8)
axs[3].legend()

# Subplot 5: Varying beta
x_values = np.linspace(-5, 5, 1000)
y_values1 = pdf_from_cf(x_values, loc=loc, scale=scale, alpha=alpha, lam=lam, beta=-1, k_max=15.0, ln_2_k_points=12, x_lim=250)
y_values2 = pdf_from_cf(x_values, loc=loc, scale=scale, alpha=alpha, lam=lam, beta=0, k_max=15.0, ln_2_k_points=12, x_lim=250)
y_values3 = pdf_from_cf(x_values, loc=loc, scale=scale, alpha=alpha, lam=lam, beta=1, k_max=15.0, ln_2_k_points=12, x_lim=250)
axs[4].plot(x_values, y_values1, label='beta=-1', color='#C00000', linewidth=1.00)
axs[4].plot(x_values, y_values2, label='beta=0', color='#000000', linewidth=1.00)
axs[4].plot(x_values, y_values3, label='beta=1', color='#008000', linewidth=1.00)
axs[4].set_title("Varying beta")
axs[4].grid(color='#E8E8E9', linestyle='-', linewidth=0.8)
axs[4].legend()

plt.tight_layout()
plt.show()
```

```python
# Default parameters
loc = 2
scale = 5
alpha = 0.7
lam = 2
beta = -0.8
```

```python
# Generate k values from -15 to 15 with 1000 points
k_values = np.linspace(-5, 5, 5001)

# Evaluate the function for each k value
y_values = np.exp(-_psi_tld(k_values, loc, scale, alpha, lam, beta))

# Split real and imaginary parts
real_part = np.real(y_values)
imag_part = np.imag(y_values)

# Plot real and imaginary parts separately
plt.figure(figsize=(12, 6))
# Plot real and imaginary parts in the same plot
plt.plot(k_values, real_part, color='#C00000',label='Real part')
plt.plot(k_values, imag_part,color='#000000', label='Imaginary part')
plt.xlabel(r'$k$')
plt.ylabel(r'$\exp(-\psi_{TLD}(k))$')
plt.title('Plot of the Characteristic Function')
plt.legend()
plt.grid(color='#E8E8E9', linestyle='-', linewidth=0.8)
plt.show()
```
